```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For better plot visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
df = sns.load_dataset('titanic')
```

```python
print("First 5 rows of dataset:")
display(df.head())
```

```python
print("\nDataset Info:")
df.info()
```

```python
print("\nStatistical Summary (numerical columns):")
display(df.describe())
```

```python
print("\nValue counts for 'Survived':")
print(df['survived'].value_counts())
```

```python
print("\nMissing values per column:")
print(df.isnull().sum())
```

```python

sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Histogram for numerical features
df.hist(bins=30, figsize=(15, 10), color='skyblue')
plt.suptitle("Histograms of Numerical Columns", fontsize=16)
plt.show()
```

```python
# Survival Count
sns.countplot(x='survived', data=df, palette="pastel")
plt.title("Survival Count")
plt.show()
```

```python
# Passenger Class Count
sns.countplot(x='pclass', data=df, palette="Set2")
plt.title("Passenger Class Count")
plt.show()
```

```python
# Survival by Gender
sns.countplot(x='sex', hue='survived', data=df, palette="husl")
plt.title("Survival Count by Gender")
plt.show()
```

```python
# Survival by Passenger Class
sns.countplot(x='pclass', hue='survived', data=df, palette="coolwarm")
plt.title("Survival Count by Passenger Class")
plt.show()
```

```python
# Age distribution by Survival
sns.boxplot(x='survived', y='age', data=df, palette="muted")
plt.title("Age Distribution by Survival")
plt.show()
```

```python
pairplot_cols = ['survived', 'pclass', 'age', 'fare']
sns.pairplot(df[pairplot_cols].dropna(), hue='survived', palette='husl')
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()
```

```python
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
